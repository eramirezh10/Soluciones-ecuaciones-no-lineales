Los 3 códigos siguientes muestran la ejecución de los métodos por separado con sus respectivas iteraciones y gráfica de convergencia:

#Método de Bisección:

```python
#MÉTODO DE BISECCIÓN (Modificado)

import numpy as np
import math as math
import matplotlib.pyplot as plt

def funcion(x):
  #Funcion que se desea encontrar sus raíces
  funcion = 1/np.exp(x) - np.cos(5*x) - 6
  return funcion

def biseccion(a, b, tol, num_iteraciones, funcion):

  i = 1
  FA = funcion(a)
  p_0 = b

  error = []  # Lista vacia para guardar el error relativo en cada iteracion
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:

    p = a + ((b-a)/2) #Criterio para "p" en este método
    FP = funcion(p)

    error_relativo = np.abs((p - p_0) / p)  #Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    print(f"Se realizaron {i} iteraciones")
    print(f"\nIntervalo = [{a}, {b}]")
    print("\np =", a + ((b-a)/2))
    print("f(a) =", funcion(a))
    print("f(b) =", funcion(b))
    print("f(p) =", funcion(p))
    print("Error relativo:", error_relativo)

    print("\n")

    if FP == 0 or ((b-a)/2) < tol:
      plt.xlabel("Iteraciones [i]", fontsize=12)
      plt.ylabel("Error relativo [%]", fontsize=12)
      plt.plot(iteraciones, error,"o-", color="red", ls=":", label = "Bisección")
      plt.title("Método de Bisección (Convergencia)", fontsize=15)
      plt.legend()
      plt.grid()
      return print(f"RESULTADO = {p}")

    i += 1

    if FA*FP > 0:
      a = p
      FA = FP

    else:
      b = p

    p_0 = p

  print(f"El método fracasó después de {i-1} iteraciones")
  return iteraciones, error

biseccion(-5, 2, 1e-5, 100, funcion)
```

#Método de Regla Falsa:

```python
#MÉTODO DE REGLA FALSA (Modificado)

import numpy as np
import math as math
import matplotlib.pyplot as plt

def funcion(x):
  #Funcion que se desea encontrar sus raíces
  funcion = 1/np.exp(x) - np.cos(5*x) - 6
  return funcion

def posicion_falsa(a, b, tol, num_iteraciones, funcion):

  i = 1
  FA = funcion(a)
  p_0 = b

  error = []  # Lista vacia para guardar el error relativo en cada iteracion
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:

    p = b - (funcion(b)*(a-b))/(funcion(a)-funcion(b)) #Criterio para "p" en este método
    FP = funcion(p)

    error_relativo = np.abs((p - p_0) / p)  # Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    print(f"Se realizaron {i} iteraciones")
    print(f"\nIntervalo = [{a}, {b}]")
    print("\np =", b - (funcion(b)*(a-b))/(funcion(a)-funcion(b)))
    print("f(a) =", funcion(a))
    print("f(b) =", funcion(b))
    print("f(p) =", funcion(p))
    print("Error relativo:", error_relativo)

    print("\n")

    if FP == 0 or np.abs((p - p_0) / p) < tol:
      plt.xlabel("Iteraciones [i]", fontsize=12)
      plt.ylabel("Error relativo [%]", fontsize=12)
      plt.plot(iteraciones, error,"o-", color="green", ls=":", label = "Regla Falsa")
      plt.title("Método de Regla Falsa (Convergencia)", fontsize=15)
      plt.legend()
      plt.grid()
      return print(f"RESULTADO = {p}")

    i += 1

    if FA*FP > 0:
      a = p
      FA = FP

    else:
      b = p

    p_0 = p

  print(f"El método fracasó después de {i-1} iteraciones")
  return iteraciones, error

posicion_falsa(-5, 0, 1e-5, 100, funcion)
```

#Método de Newton-Raphson:

```python
#MÉTODO DE NEWTON-RAPHSON (Modificado)

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def funcion_derivada():
  x = sp.symbols('x')

  funcion_simb = 1/sp.exp(x) - sp.cos(5*x) - 6
  derivada_simb = sp.diff(funcion_simb, x)

  funcion_num = sp.lambdify(x, funcion_simb)
  derivada_num = sp.lambdify(x, derivada_simb)

  return funcion_num, derivada_num

def newton(p_0, tol, num_iteraciones, funcion_derivada):
  funcion_num, derivada_num = funcion_derivada()

  f_x = funcion_num
  d_x = derivada_num

  i = 1
  error = []  # Lista vacia para guardar el error relativo en cada iteracion
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:
    p = p_0 - f_x(p_0) / d_x(p_0)  # Condición

    error_relativo = np.abs((p - p_0) / p)  # Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    print(f"Se realizaron {i} iteraciones:")
    print("p =", p_0)
    print("f(p) =", f_x(p_0))
    print("f'(p) =", d_x(p_0))
    print("Error relativo =", error_relativo)
    print("\n")

    if np.abs(p - p_0) < tol:
      plt.xlabel("Iteraciones [i]", fontsize=12)
      plt.ylabel("Error relativo [%]", fontsize=12)
      plt.plot(iteraciones, error,"o-", color="red", ls=":", label = "Newton-Raphson")
      plt.title("Método de Newton-Raphson (Convergencia)", fontsize=15)
      plt.legend()
      plt.grid()
      return print(f"\nRESULTADO = {p}")

    else:
      i += 1
      p_0 = p

  print(f"El método fracasó después de {i-1} iteraciones")
  return iteraciones, error

newton(-1.5, 1e-5, 20, funcion_derivada)
```

#Convergencia de los métodos para encontrar raíces.

Ahora, el siguiente código muestra los 3 métodos en un solo programa que grafica la convergencia de estos en una sola gráfica de convergencia:

```python
#GRAFICACIÓN DE CONVERGENCIA ENTRE MÉTODOS

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def funcion(x):
  f = 1/np.exp(x) - np.cos(5*x) - 6
  return f

def funcion_derivada():
  x = sp.symbols('x')

  funcion_simb = 1/sp.exp(x) - sp.cos(5*x) - 6
  derivada_simb = sp.diff(funcion_simb, x)

  funcion_num = sp.lambdify(x, funcion_simb)
  derivada_num = sp.lambdify(x, derivada_simb)

  return funcion_num, derivada_num

#-----------------------------------------------------------------------------------------------------------

#MÉTODO DE BISECCIÓN

def biseccion(a, b, tol, num_iteraciones, funcion):
  i = 1
  FA = funcion(a)
  p_0 = b

  error = []  # Lista vacía para guardar el error relativo en cada iteración
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:

    p = a + ((b-a)/2) # Criterio para "p" en este método
    FP = funcion(p)

    error_relativo = np.abs((p - p_0) / p)  # Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    if FP == 0 or ((b-a)/2) < tol:
      break

    i += 1

    if FA*FP > 0:
      a = p
      FA = FP
    else:
      b = p

    p_0 = p

  return iteraciones, error

#-----------------------------------------------------------------------------------------------------------

#REGLA FALSA

def posicion_falsa(a, b, tol, num_iteraciones, funcion):
  i = 1
  FA = funcion(a)
  p_0 = b

  error = []  # Lista vacia para guardar el error relativo en cada iteracion
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:

    p = b - (funcion(b)*(a-b))/(funcion(a)-funcion(b)) #Criterio para "p" en este método
    FP = funcion(p)

    error_relativo = np.abs((p - p_0) / p)  # Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    if FP == 0 or np.abs((p - p_0) / p) < tol:
      break

    i += 1

    if FA*FP > 0:
      a = p
      FA = FP

    else:
      b = p

    p_0 = p

  return iteraciones, error


#-----------------------------------------------------------------------------------------------------------

#MÉTODO DE NEWTON

def newton(p_0, tol, num_iteraciones, funcion_derivada):
  funcion_num, derivada_num = funcion_derivada()

  i = 1
  error = []  # Lista vacía para guardar el error relativo en cada iteración
  iteraciones = []  # Lista para almacenar el número de iteraciones

  while i <= num_iteraciones:

    p = p_0 - funcion_num(p_0) / derivada_num(p_0)  # Condición

    error_relativo = np.abs((p - p_0) / p)  # Calcula el error relativo

    error.append(error_relativo)
    iteraciones.append(i)

    if np.abs(p - p_0) < tol:
      break

    else:
      i += 1
      p_0 = p

  return iteraciones, error

#-----------------------------------------------------------------------------------------------------------

#EJECUCIÓN
i_biseccion, e_biseccion = biseccion(-5, 2, 1e-5, 100, funcion)
i_rfalsa, e_rfalsa = posicion_falsa(-5, 0, 1e-5, 100, funcion)
i_newton, e_newton = newton(-1.5, 1e-5, 100, funcion_derivada)

#GRAFICAR CONVERGENCIA

plt.xlim(0,45) #Límites del eje x
plt.ylim(0,2.5) #Límites del eje y

plt.xlabel("Iteraciones [i]", fontsize=12)
plt.ylabel("Error relativo [%]", fontsize=12)

plt.plot(i_biseccion, e_biseccion, "o-", color="red", ls=":", label = "Bisección")
plt.plot(i_rfalsa, e_rfalsa, "o-", color="green", ls=":", label = "Regla Falsa")
plt.plot(i_newton, e_newton, "o-", color="blue", ls=":", label = "Newton")

plt.title("Convergencia entre métodos", fontsize=15)
plt.legend()
plt.grid()
```
